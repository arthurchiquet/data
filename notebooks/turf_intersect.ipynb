```python
import json
import pandas as pd
from turfpy.transformation import intersect, union
from geojson import FeatureCollection
```

```python
df = pd.read_csv('communes_aoc.csv', sep=";")
df["CI"] = (
        df["CI"].astype(str).map(lambda x: (5 - len(x)) * "0" + x)
    )
liste_communes = df.CI.to_list()
```

```python
import requests

features = []
for c in liste_communes:
  url = f"https://cadastre.data.gouv.fr/bundler/cadastre-etalab/communes/{c}/geojson/communes"
  response = requests.get(url)
  try:
    features.append(response.json()["features"][0])
  except:
    print(c)
    print(response.json())
```

```python
import requests

url_1 = "https://raw.githubusercontent.com/arthurchiquet/data/refs/heads/main/vineyard.geojson"
url_2 = "https://raw.githubusercontent.com/arthurchiquet/data/refs/heads/main/communes_aoc.geojson"

response_1 = requests.get(url_1)
response_2 = requests.get(url_2)
aoc = response_1.json()
ld = response_2.json()
```

```python
for f in ld['features']:
  f["id"] = f["properties"]["code"]
```

```python
import json
with open("communes_aoc.geojson", "w") as file:
    json.dump(ld, file)
```

```python
new_aoc = union(aoc)
```

```python
new_features = []
for f in ld['features']:
  it = intersect([f, new_aoc])
  try:
    if it["geometry"]["coordinates"]!= [[]]:
      new_features.append(it)
  except:
    print(f)
```

```python
new_geojson = FeatureCollection(new_features)
```

```python
new_features = []
for f in aoc['features']:
  for g in ld['features']:
    it = intersect([f, g])
    try:
      if it["geometry"]["coordinates"]!= [[]]:
        new_features.append(it)
    except:
      print(g)
```

```python
new_geojson = FeatureCollection(new_features)
```
